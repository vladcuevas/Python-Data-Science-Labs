## Segment 1 - Creating standard data graphics

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

### Creating a line chart from a list object

#### Plotting a line chart in matplotlib

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
```

#### Plotting a line chart from a Pandas object

```python
import pathlib
import os
address = pathlib.Path(os.getcwd()).parent
address = pathlib.Path(os.path.join(address, 'Data/mtcars.csv'))
```

```python
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars['mpg']
```

```python
mpg.plot()
```

```python
df = cars[['cyl','wt','mpg']]
df.plot()
```

### Creating bar charts
#### Creating a bar chart from a list

```python
plt.bar(x, y)
```

#### Creating bar charts from Pandas objects

```python
mpg.plot(kind="bar")
```

```python
mpg.plot(kind="barh")
```

### Creating a pie chart

```python
x = [1,2,3,4,0.5]
plt.pie(x)
plt.show()
```

### Saving a plot

```python
plt.pie(x)
plt.savefig('pie_chart.png')
plt.show()
```
